```python
pd.read_csv("citibike_weather_2022.csv", nrows=2).columns
```

```python

# Read final merged dataset
df = pd.read_csv("citibike_weather_2022.csv", parse_dates=["date"])

# Quick check
print(df.head())
print(df.columns)
```

## Line plot of daily max temperature (TMAX)

```python
df.plot(x="date", y="TMAX", figsize=(12,5), title="Daily Max Temperature (°C) in 2022")
plt.ylabel("Temperature (°C)")
plt.show()
```

## Trips per day

```python
print(df[["date", "trips"]].head())
```

## Dual-axis line chart: trips vs. temperature

```python
fig, ax1 = plt.subplots(figsize=(12,5))

# Left axis: trips
ax1.plot(df["date"], df["trips"], color="tab:blue", label="Trips")
ax1.set_ylabel("Number of Trips", color="tab:blue")

# Right axis: temperature
ax2 = ax1.twinx()
ax2.plot(df["date"], df["TMAX"], color="tab:red", label="Max Temp (°C)")
ax2.set_ylabel("Temperature (°C)", color="tab:red")

plt.title("CitiBike Trips vs. Daily Max Temperature (2022)")
fig.tight_layout()
plt.show()
```

## I used Matplotlib (via pandas `.plot` and object-oriented `ax1/ax2`) to visualize the data.
- First, a simple time series of TMAX.
- Then, a dual-axis chart combining trips (left y-axis) and temperature (right y-axis).

This uses Matplotlib’s object-oriented paradigm, where we create `ax1` and then a twin axis (`ax2`) to overlay different scales.
